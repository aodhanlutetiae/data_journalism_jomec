## colouring dfs

```python
# import cardiff weather

path = '/Users/aidanair/Documents/DATA/ALL_DATASETS/'
file = 'cardiff_weather.csv'

df = pd.read_csv(path + file)
```

```python
# filter for last year

last = df[df.Year == 2019]
```

```python
# sort by rainfall

last.sort_values(by = 'Rain (mm)')
```

```python
# runa a bar chart in each column

last.sort_values(by = 'Rain (mm)', ascending = False).style.bar(color= 'lightblue')
```

```python
# create a function to filter the numerical value into red color (ignoring strings an integers)

def color_below_20_red(value):
    
    if type(value) == type(''):
        return 'color:black'
    elif type(value) == int:
        return 'color:black'
    else:
        color = 'red' if value >= 80 else 'black'
        return 'color: {}'.format(color)
```

```python
# apply function to any float cols (with data over 80)

last.style.applymap(color_below_20_red)
```

```python
# colour on values along a scale min-max in each col

last.style.background_gradient(cmap = 'Blues')
```

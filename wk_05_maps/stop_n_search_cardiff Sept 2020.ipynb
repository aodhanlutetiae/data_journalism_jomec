> This notebook filters the data down to incidents that have lat, long data

```python
import pandas as pd
```

```python
path = '/Users/aidanair/Documents/DATA/ALL_DATASETS/'
file = '2020-09-south-wales-stop-and-search.csv'

df = pd.read_csv(path + file)
```

```python
df.shape
```

```python
# view df by date to check what we're looking at

df.sort_values(by = 'Date')
```

```python
# what kind of incidents don't have lat, long?

missing = df[df.Latitude.isnull()]
missing
```

```python
# build a new df called df_ll that excludes rows where 'Latitude' is NaN

df_ll = df[~df.Latitude.isnull()]
```

```python
# what size is it now, only with rows that have lat, long?

df_ll.shape
```

```python
# what's the male / female breakdown?

df_ll.Gender.value_counts()
```

```python
# what's the range on the lat, long - 

print(df.Latitude.min())
print(df.Latitude.max())
print(df.Longitude.min())
print(df.Longitude.max())
```

```python
# filter two dataframes - one for males stopped and one for females

male_ss_ll_092020 = df_ll[df_ll.Gender == 'Male']

female_ss_ll_092020 = df_ll[df_ll.Gender == 'Female']
```

```python
# what size is each dataframe?

print(len(male_ss_ll_092020))
print(len(female_ss_ll_092020))
```

```python
# export the two datasets

male_ss_ll_092020.to_csv('male_ss_sept.csv', index = False) 
female_ss_ll_092020.to_csv('female_ss_sept.csv', index = False) 
```

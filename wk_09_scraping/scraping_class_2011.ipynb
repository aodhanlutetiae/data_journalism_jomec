```python
# assemble via a list of lists - tighter version

big_list = []

for a in soup.find_all("div", class_ = "profile search-result with-image vcard"):
    
    small_list = []
    
    for b in a.find_all("h2"):
        name = b.text.strip('\n')

    for c in a.find_all("p"):
        job = c.text

    for d in a.find_all("dd", class_ = 'profile-contact-telephone'):
        phone = d.text
        
    for e in a.find_all('dd'):
        for f in e.find_all('a', href=True):
            mail = f.text
            
    small_list.extend((name, job, phone, mail))
    
    big_list.append(small_list)

len(big_list)
```

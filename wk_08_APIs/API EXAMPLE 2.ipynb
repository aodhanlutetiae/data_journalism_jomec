## All stop and search around a point (lat, long) in a given month

```
endpoint = 'https://data.police.uk/api/stops-street?'
parameters = {'lat':"52.629729", 'lng':"-1.131592", 'date':"2020-10"}

req = requests.get(endpoint, params = parameters)
print(req)
response = req.json()
```

## Query and get data: All bike theft around a point in a given month

```
endpoint = 'https://data.police.uk/api/crimes-street/bicycle-theft?'
parameters = {'lat':"51.476369992679096", 'lng':"-3.1781098361443667", 'date':"2020-10"}

req = requests.get(endpoint, params = parameters)
print(req)
response = req.json()
print(len(response))
```

## Assemble the data into a df

```
# can we print what we need?

for x in response:
  print(x['location']['latitude'])
  print(x['location']['longitude'])
  print(x['location']['street']['name'])
  print(x['outcome_status'])
```

```

rows = []

for x in response:

  lat = x['location']['latitude']
  long = x['location']['longitude']
  street = x['location']['street']['name']
  outcome = x['outcome_status']['category']

  rows.append({"latitude": lat, "longitude": long, "place": street, "result": outcome})
```

```
df = pd.DataFrame(rows)
df[:4]
```
